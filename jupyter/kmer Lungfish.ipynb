```python
%matplotlib inline

import matplotlib.pyplot as plt

kmers = {'k=25': '/groups/tanaka/Projects/lungfish-genome/genome_size/k25/lungfish.k25.histo',
         'k=19': '/groups/tanaka/Projects/lungfish-genome/genome_size/k19/HJK2HDMXX_merged.k19.histo',
         'k=21': '/groups/tanaka/Projects/lungfish-genome/genome_size/k21/merged.k21.histo',
         'k=23': '/groups/tanaka/Projects/lungfish-genome/genome_size/k23/merged.k23.histo',
         'k=27': '/groups/tanaka/Projects/lungfish-genome/genome_size/k27/HJK2HDMXX.k27.histo',
         'k=29': '/groups/tanaka/Projects/lungfish-genome/genome_size/k29/HJK2HDMXX.k29.histo',
         'k=31': '/groups/tanaka/Projects/lungfish-genome/genome_size/k31/HJK2HDMXX.k31.histo'}

plt.figure(figsize=(20,10))
for i, k in enumerate(sorted(kmers)):
    data_x = []
    data_y = []
    
    print(f'Reading the file {kmers[k]}')
    with(open(kmers[k], 'r')) as hFile:
        iSum = 0
        for line in hFile.readlines():
            x, y = line.strip().split(' ')
            data_x.append(int(x))
            data_y.append(int(y))
            iSum += int(x) * int(y)
            
    #print(f'SUM: {sum}')
    #data_y[:] = [(x / sum) * 100 for x in data_y]
    tmp_x = data_x[101:]
    tmp_y = data_y[101:]
    tmp_sum = sum(tmp_y)
    data_y[101] = tmp_sum
    print(data_y[95:105])
    print(tmp_sum)
    #plt.plot(data_x[2:102], data_y[2:102], label = k)
    plt.plot(data_x[2:100], data_y[2:100], label = k)

plt.legend(fontsize=20)
yloc, ylab = plt.yticks()
ylab_new = []
for v,l in zip(yloc, ylab):
    if v / 1_000_000 > 1:
        ylab_new.append(f'{v / 1_000_000}M')
    elif v / 1_000 > 1:
        ylab_new.append(f'{v / 1_000}K')
    else:
        ylab_new.append(f'{v:.2f}')
plt.yticks(yloc, ylab_new)
for tick in plt.gca().xaxis.get_major_ticks():
    tick.label.set_fontsize(20) 
for tick in plt.gca().yaxis.get_major_ticks():
    tick.label.set_fontsize(20) 

plt.ylim([0, 1.2e9])
#plt.ylim([0, 0.12])
plt.xlabel('k-mer abundance', fontsize=24)
plt.ylabel('Number of k-mers', fontsize=24)
plt.grid()
plt.title('Spectrum of read k-mer abundance\n', fontsize=30)
plt.savefig('/groups/tanaka/Projects/lungfish-genome/genome_size/kmer19-31.pdf') 
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import statistics as stat

data = []
#with(open('/groups/tanaka/Projects/axolotl-genome/AmexG_v6.0/AmexG_v6.0_DD/work/manuscript/compare_Bionano_with_Marvel/nums', 'r')) as hFile:
with(open('/groups/tanaka/Projects/lungfish-genome/transcriptome_Thorsten/contamination_check/still_unmapped.len.dist', 'r')) as hFile:
    for line in hFile.readlines():
        data.append(int(line.strip().split('\t')[1]))

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.hist(data, bins=100)
mn = stat.mean(data)
md = stat.median(data)
plt.axvline(mn, color='r')
plt.axvline(md, color='g')
plt.text(mn,13700,f'Mean: {mn:.2f}', fontsize=20)
plt.text(md,12000,f'Median: {md:.2f}', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.xlabel('Contig length', fontsize=20)
plt.grid()

plt.subplot(1,2,2)
plt.hist(data, bins=1000)
mn = stat.mean(data)
md = stat.median(data)
plt.axvline(mn, color='r')
plt.axvline(md, color='g')
plt.text(mn,2000,f'Mean: {mn:.2f}', fontsize=20)
plt.text(md,1500,f'Median: {md:.2f}', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.xlabel('Contig length', fontsize=20)
plt.xlim([0,1000])
plt.grid()
plt.savefig('/groups/tanaka/Projects/lungfish-genome/transcriptome_Thorsten/contamination_check/hist_still_unmapped.pdf') 
```
