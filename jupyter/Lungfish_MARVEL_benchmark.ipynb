```python
%matplotlib inline

import matplotlib.pyplot as plt
import sys
import statistics as stats

fname = '/groups/tanaka/Projects/lungfish-genome/revision/MKL2_read_lengths.txt'

lengths = []
with(open(fname, 'r')) as hFile:
    for line in hFile.readlines():
        lengths.append(int(line.strip()))

```

```python
avg = sum(lengths) / len(lengths)
print(f'Mean: {avg}')
med = stats.median(lengths)
print(f'Median: {med}')
sd = stats.stdev(lengths)
print(f'SD: {sd}')

lengths_n50 = lengths
lengths_n50.sort()
newlist = []
for x in lengths_n50 :
    newlist += [x]*x

nElem = len(newlist)
n50 = 0
if nElem % 2 == 0:
    n50 = (newlist[int(nElem / 2) - 1] + newlist[int(nElem / 2)]) / 2
else:
    n50 = newlist[int(nElem / 2)]
print(f'N50: {n50}')
```

```python
lengths2 = [min(i, 50_000) for i in lengths]
print(lengths2[1:10])
plt.figure(figsize=(20,10))
plt.hist(lengths2, bins=1000)
plt.axvline(x=avg, color='red')
#plt.xlim([0, 500000])
plt.xlabel('Read length, bp', fontsize=21)
plt.ylabel('Count', fontsize=21)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```
